# The SIR model

```python
%matplotlib notebook
%load_ext autoreload
%autoreload 2

from sir_model import SIR
```

\begin{split}
&dS(t)=-\beta I(t) \frac{S(t)}{N}\\
&dI(t) = \beta I(t) - \gamma I(t)\\
&dR(t) = (1-p_d) \gamma I(t)\\
&dD(t) = p_d \gamma I(t)
\end{split}

$R_0 = \frac{\beta}{\gamma}$, $\gamma=1/14$, $R_0=2$

```python
model = SIR(param={
    'beta': 2 / 14, 
    'gamma': 1 / 14,
    'p_d': 0.1})
model.run(
    X_init=[0.9, 0.1, 0, 0], 
    num_steps=300)
model.make_plot()
```
