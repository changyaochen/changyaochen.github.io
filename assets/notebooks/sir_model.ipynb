# The SIR model

```python
%matplotlib notebook
%load_ext autoreload
%autoreload 2

from sir_model import SIR
```

\begin{split}
\text{d}S(t) &= -\beta I(t) \frac{S(t)}{S(t) + I(t) + R(t)}\\
\text{d}I(t) &= -\text{d}S(t) - \gamma I(t)\\
\text{d}R(t) &= (1-\eta) \gamma I(t)\\
\text{d}D(t) &= \eta \gamma I(t)
\end{split}

Parameter example: $\gamma=1/14$, $R_0=2$, $R_0 = \frac{\beta}{\gamma}$,

```python
model = SIR(param={
    'beta': 3 / 14, 
    'gamma': 1 / 14,
    'eta': 0.02})
model.run(
    X_init=[0.9, 0.1, 0, 0], 
    num_steps=120)
model.make_plot()
```
